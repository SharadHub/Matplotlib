Data visualization that shows relationships between two variables

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

Comparing two datasets from 2 different days

```python
# Draw two plots on the same figure:

import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
import pandas as pd

df_marketing = pd.DataFrame({
    'Region': ['North', 'South', 'East', 'West', 'Central', 'Northwest', 'Southeast', 'Southwest'],
    'Marketing_Spend': [20, 25, 15, 30, 10, 18, 22, 27],  # in $1,000
    'Sales_Revenue': [180, 200, 120, 250, 100, 150, 170, 230]  # in $1,000
})


df_study = pd.DataFrame({
    'Student_ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A006', 'A007', 'A008'],
    'Hours_Studied': [1, 2, 4, 3, 5, 6, 8, 7],
    'Exam_Score': [45, 48, 65, 60, 70, 80, 90, 85]  # in percentage
})

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.scatter(df_marketing['Marketing_Spend'],df_marketing['Sales_Revenue'], color='cyan', edgecolors='black')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales Revenue')
plt.title('Marketing v/s Sales in various Region')

plt.subplot(1,2,2)
plt.scatter(df_study['Hours_Studied'], df_study['Exam_Score'], color='orange', edgecolors='black')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score(in %)')
plt.title('Hour Studied v/s Score Board')

plt.suptitle('Comparison board')
plt.tight_layout()
plt.show()
```
