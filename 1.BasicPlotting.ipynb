```python
# Example 1: Basic Line Plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# Customized Line Plot

plt.plot(x, y, color='green', linestyle='--', marker='o', label='Square')
plt.title("Styled Line Plot")
plt.xlabel("Temperature")
plt.ylabel("Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 4, 6, 8]

plt.plot(x, y1, label="Squares", color='blue')
plt.plot(x, y2, label="Doubles", color='orange')
plt.title("Multiple Lines")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Plot the cube of numbers [1,2,3,4,5]

x = [1, 2, 3, 4, 5]
y = [i**3 for i in x]

plt.plot(x, y)
plt.title("Cube Plot")
plt.grid()
plt.show()
```

```python
# Add a red dashed line with circle markers

plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.show()
```
