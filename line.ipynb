```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

there are many ways to make marker like:
marker = 'o', 'v', '*' etc

To make line we use:
linestyle = '-', ':', '--', '-.' etc

```python
plt.plot(ypoints, marker = '*', linestyle=':')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', linestyle='-.', ms=10, mec='hotpink', mfc='hotpink')
plt.show()
```
