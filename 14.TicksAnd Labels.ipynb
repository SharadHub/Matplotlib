Ticks are the values shown on the x-axis and y-axis to indicate scale. By customizing them, you can improve readability (e.g., rotate long labels)

Highlight specific intervals or formats (e.g., 10%, 20%, 30% or Jan, Feb, Mar)

Handle categories or datetime formats

Avoid clutter in tight spaces

Syntax:

plt.xticks(ticks=None, labels=None, rotation=0, fontsize=None)
plt.yticks(ticks=None, labels=None, rotation=0, fontsize=None)


```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [10, 20 , 30]

plt.plot(x, y)
plt.xticks([1, 2, 3], ['Q1', 'Q2', 'Q3'])
plt.title("Quarter Wise Growth")
plt.show()
```

```python
month = ['Januray', 'Februrary', 'March', 'April']
sales = [10500, 11000, 9500, 11500]

plt.plot(month, sales)
plt.xticks(rotation = 45)
plt.title("Monthly Sales")
plt.show()
```

```python
plt.plot([0, 1, 2], [100, 200, 300])
plt.yticks([100, 200, 300], ['Low', 'Medium', 'High'])
plt.title("Data Visualization")
plt.show()
```
