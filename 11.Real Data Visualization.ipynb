```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('austin_weather.csv')
df['Date'] = pd.to_datetime(df['Date'])  # parse date

plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['TempHighF'], label="High Temp", color='red')
plt.xlabel("Date")
plt.ylabel("Temperature (F)")
plt.title("Austin High Temperature Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month')['TempHighF'].mean()

plt.figure(figsize=(10,6))
monthly_avg.plot(kind='bar', cmap='coolwarm')
plt.title("Average High Temp by Month")
plt.ylabel('Temperature (F)')
plt.annotate('Hot Day!',
             xy=(df['Date'].iloc[150], df['TempHighF'].iloc[150]),
             xytext=(df['Date'].iloc[130], 105),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('austin_weather.csv')
df['Date'] = pd.to_datetime(df['Date'])

# Bar: Average temperature by month
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month')['TempHighF'].mean()

# Histogram: distribution of daily high temps
temps = df['TempHighF'].dropna()

fig, axs = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Bar Chart
axs[0].bar(monthly_avg.index, monthly_avg.values, color='skyblue')
axs[0].set_title("Avg High Temp by Month")
axs[0].set_xlabel("Month")
axs[0].set_ylabel("Temp (F)")

# Histogram
axs[1].hist(temps, bins=20, color='salmon', edgecolor='black')
axs[1].set_title("High Temp Distribution")
axs[1].set_xlabel("Temp (F)")
axs[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
