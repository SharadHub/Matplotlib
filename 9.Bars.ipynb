In Pyplot, we can draw bars by using bar() function

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
# Example 1: Basic Vertical Bar Plot

products = ['Laptop', 'Tablet', 'Smartphone']
sales = [120, 90, 160]

plt.bar(products, sales, color='orange')
plt.title('Product Sales')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.show()
```

```python
#  Example 2: Horizontal Bar Chart

plt.barh(products, sales, color='green')
plt.title('Product Sales (Horizontal)')
plt.xlabel('Units Sold')
plt.ylabel('Product')
plt.show()
```

```python
# Example 3: Sorted Bars with Colors

import numpy as np

categories = ['B', 'A', 'C']
values = [20, 10, 30]
sorted_idx = np.argsort(values)

plt.bar(np.array(categories)[sorted_idx], np.array(values)[sorted_idx], color='tomato')
plt.title('Sorted Bar Plot')
plt.show()
```

```python
# Example 4: Bar Widths and Custom Colors

colors = ['#FF9999', '#66B2FF', '#99FF99']
plt.bar(categories, values, color=colors, width=0.5)
plt.title('Custom Colors & Width')
plt.show()
```

```python
# Example 5: Adding Data Labels

bars = plt.bar(categories, values, color='purple')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, str(yval), ha='center', va='bottom')

plt.title('Bar Plot with Labels')
plt.show()
```

1. Plot Sales for 5 Products.
Input: Product names and sales numbers.
Goal: Create a simple bar chart.

```python
data = {
    'Product' : ['Book', 'Textbook', 'Pen', 'Headphone', 'Mobile'],
    'Sales_number' : [100, 500, 80, 20, 1100]
}
import pandas as pd
df = pd.DataFrame(data)
df = df.sort_values(by='Sales_number')

plt.bar(df['Product'],df['Sales_number'], color='tomato')
plt.title('Product Sold')
plt.xlabel('Products')
plt.ylabel('Sales Number')

for index, value in enumerate(df['Sales_number']):
    plt.text(index, value + 20, str(value), ha='center')


plt.show()
```

2. Horizontal Plot of Favorite Programming Languages.
Data: ['Python', 'Java', 'C++', 'JavaScript']
Votes: [350, 300, 150, 400]

```python
programming_language = {
    'Data': ['Python', 'Java', 'C++', 'JavaScript'],
    'Votes': [350, 300, 150, 400]
}
data_frame = pd.DataFrame(programming_language)
data_frame = data_frame.sort_values(by='Votes')

plt.barh(data_frame['Data'], data_frame['Votes'], color='cyan')
plt.title('Programming Language Popularity')
plt.xlabel('Votes')
plt.ylabel('Programming Language')

for index, value in enumerate(data_frame['Votes']):
    plt.text(value + 5, index,str(value), ha='center')

plt.show()
```

3. Compare Monthly Revenue
Data: {'Jan': 1000, 'Feb': 1200, 'Mar': 800, 'Apr': 1500}

Plot with sorted bars (high to low)
Use different bar colors

```python
Data = {'Jan': 1000, 'Feb': 1200, 'Mar': 800, 'Apr': 1500}

Data1 = pd.DataFrame(Data.items(), columns=['Month', 'Sales'])
Data1 = Data1.sort_values(by='Sales', ascending=False)

colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3']

plt.bar(Data1['Month'], Data1['Sales'], color=colors, edgecolor='black')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')

for i, value in enumerate(Data1['Sales']):
    plt.text(i, value + 50, str(value), ha='center')
plt.show()
```

4. Add Data Labels Above Bars
Plot daily foot traffic in a store and display exact counts on each bar.

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
foot_traffic = [230, 180, 290, 310, 270]

plt.bar(days, foot_traffic, color='slateblue')
plt.title('Daily Foot Traffic')
plt.xlabel('Day')
plt.ylabel('People Count')

# Add data labels
for i, val in enumerate(foot_traffic):
    plt.text(i, val + 5, str(val), ha='center')

plt.show()
```

5. Dual Comparison with Side-by-Side Bars (Grouped)
Data:
months = ['Jan', 'Feb', 'Mar']
sales_2024 = [100, 200, 150]
sales_2025 = [110, 210, 180]

Goal:
Plot bars side-by-side for each month (2024 vs. 2025).
Use np.arange() and bar width logic.

```python

months = ['Jan', 'Feb', 'Mar']
sales_2024 = [100, 200, 150]
sales_2025 = [110, 210, 180]

x = np.arange(len(months))  # label locations
width = 0.35  # bar width

plt.bar(x - width/2, sales_2024, width, label='2024', color='steelblue')
plt.bar(x + width/2, sales_2025, width, label='2025', color='orange')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison: 2024 vs 2025')
plt.xticks(x, months)
plt.legend()
plt.tight_layout()
plt.show()
```
