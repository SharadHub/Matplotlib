```python
# Plot Daily Temperatures

import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temps = [30, 32, 34, 33, 31, 29, 28]

# Your task: Plot a line graph with title, markers, color, grid, and show it.

plt.plot(days, temps, marker='o', color='blue', linestyle='-', label='Temperature')
plt.title("Weekly Temperature Trend")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.legend()
plt.show()
```
