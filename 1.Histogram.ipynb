```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(0,100,100)
plt.hist(data, bins=10, color='yellow', edgecolor='black')
plt.title("Histogram Title")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

```python
#intermediate
import numpy as np

data = np.random.normal(50, 10, 200)  # 200 data points around 50
plt.hist(data, bins=10, color='orange', edgecolor='black')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#hard
data1 = np.random.normal(50, 5, 100)
data2 = np.random.normal(70, 10, 100)

plt.hist([data1, data2], bins=10, color=['blue', 'red'], label=['Group A', 'Group B'])
plt.legend()
plt.title("Comparison of Two Groups")
plt.show()
```

# Generate 300 random values with mean 60 and std 8. Show histogram with 15 bins and red bars.

```python
random_data = np.random.normal(60, 8, 300)

plt.hist(random_data, bins = 15, color = 'red', edgecolor = 'black')
plt.title("Random Data Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Compare sales distribution of 2 stores over 200 days using 10 bins. Add legends.

```python
data1 = np.random.randint(0,10000,200)
data2 = np.random.randint(0,10000,200)

plt.hist([data1,data2], bins = 10, color = ['blue', 'green'], edgecolor = 'black', label = ['Data 1', 'Data 2'])
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Days")
plt.show()
```

# Mini Project: Student Performance Distribution

Generate random scores of 200 students (range 0100).

Create histogram with 10 bins.

Color bars green with black edges.

Add proper title, x and y labels.

Analyze which score range has the most students.

```python
import numpy as np
import matplotlib.pyplot as plt

scores = np.random.randint(0, 101, 200)
plt.hist(scores, bins=10, color='green', edgecolor='black')
plt.title("Student Score Distribution")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.show()
```
